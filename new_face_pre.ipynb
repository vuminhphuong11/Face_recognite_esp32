```python
import os
from os import listdir
from PIL import Image
from numpy import asarray
from numpy import expand_dims
from matplotlib import pyplot
from keras.models import load_model
import numpy as np
import csv
import pickle
import cv2
from keras_facenet import FaceNet
```

```python
#HaarCascade = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
HaarCascade = cv2.CascadeClassifier(cv2.samples.findFile(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml'))# tải file từ opencv

#MyFaceNet = load_model('facenet_keras.h5')
MyFaceNet = FaceNet()tải file 
```

```python
main_folder = 'image_folder/'  # Thư mục chính chứa các thư mục con
output_csv = 'Attendance.csv'
database = {}

# Hàm kiểm tra trạng thái (giả định bạn có mô hình hoặc logic để xác định trạng thái)
def check_presence(signature):
    # Bạn có thể thay đổi logic tùy thuộc vào bài toán thực tế
    return "absent"

# Tạo và ghi thông tin vào file CSV
with open(output_csv, mode='w', newline='', encoding='utf-8') as csvfile:
    csv_writer = csv.writer(csvfile)
    # Ghi tiêu đề cột
    csv_writer.writerow(['Name', 'Status'])

    # Duyệt qua từng thư mục con trong thư mục chính
    for person_folder in os.listdir(main_folder):
        person_path = os.path.join(main_folder, person_folder)

        # Kiểm tra xem đó có phải là thư mục không
        if os.path.isdir(person_path):
            for filename in os.listdir(person_path):
                file_path = os.path.join(person_path, filename)

                # Đọc ảnh
                gbr1 = cv2.imread(file_path)

                # Phát hiện khuôn mặt
                the_face = HaarCascade.detectMultiScale(gbr1, 1.1, 4)
                if len(the_face) > 0:
                    x1, y1, width, height = the_face[0]
                else:
                    x1, y1, width, height = 1, 1, 10, 10

                # Xử lý tọa độ
                x1, y1 = abs(x1), abs(y1)
                x2, y2 = x1 + width, y1 + height

                # Chuyển đổi ảnh
                gbr = cv2.cvtColor(gbr1, cv2.COLOR_BGR2RGB)
                gbr = Image.fromarray(gbr)  # Chuyển từ OpenCV sang PIL
                gbr_array = asarray(gbr)

                # Cắt khuôn mặt
                face = gbr_array[y1:y2, x1:x2]
                face = Image.fromarray(face)
                face = face.resize((160, 160))
                face = asarray(face)

                # Chuẩn bị dữ liệu cho mô hình
                face = expand_dims(face, axis=0)
                signature = MyFaceNet.embeddings(face)

                # Thêm vào cơ sở dữ liệu
                database.setdefault(person_folder, []).append(signature)

            # Xác định trạng thái (có mặt/vắng mặt)
            status = check_presence(None)  # Mặc định là "asbent"

            # Ghi vào file CSV
            csv_writer.writerow([person_folder, status])

print(f"Dữ liệu đã được lưu vào {output_csv}")
```
